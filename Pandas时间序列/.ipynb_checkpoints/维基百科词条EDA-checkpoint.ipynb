```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

```

```python
train = pd.read_csv('train_1.csv').fillna(0)
train.head()
```

```python
train.info()
```

### 将 csv中所有的 float类型的数据转换成 integer 类型, 因为 float类型所占内存空间比 integer多

```python
for col in train.columns[1:]:
    train[col] = pd.to_numeric(train[col], downcast='integer')
train.head()    
```

```python
train.info()
```

**注:** 将 float 转成 integer后，所占内存空间明显得到了减少！

从以上 "Page" 列可以看出 Page 信息其实是包含了国家信息的 "2NE1_zh.wikipedia.org_all-access_spider" 中的 "zh"代表的是 『中国』...

那么我们就可以将其中的 所有的 国家信息 找出来!

```python
import re
from collections import Counter
def get_language(page):
    res = re.search('[a-z][a-z].wikipedia.org', page)
    if res:
        return res.group()[0:2]
    else:
        return 'na'
    
train['lang'] =  train.Page.map(get_language)

print(Counter(train['lang']))
```

## 统计 以上几种Language词条频率

```python
lang_sets={}
lang_sets['en'] = train[train.lang=='en'].iloc[:,0:-1]
lang_sets['ja'] = train[train.lang=='ja'].iloc[:,0:-1]
lang_sets['de'] = train[train.lang=='de'].iloc[:,0:-1]
lang_sets['na'] = train[train.lang=='na'].iloc[:,0:-1]
lang_sets['fr'] = train[train.lang=='fr'].iloc[:,0:-1]
lang_sets['zh'] = train[train.lang=='zh'].iloc[:,0:-1]
lang_sets['ru'] = train[train.lang=='ru'].iloc[:,0:-1]
lang_sets['es'] = train[train.lang=='es'].iloc[:,0:-1]

sums = {}

for key in lang_sets:
    # axis=0表述列 
    # axis=1表述行
    sums[key] = lang_sets[key].iloc[:,1:].sum(axis=0) / lang_sets[key].shape[0]
      
```

```python
days = [d for d in range(sums['en'].shape[0])]

fig = plt.figure(figsize=(18,10))
fig.add_subplot(111)
plt.ylabel('Views per Page')
plt.xlabel('Days')

plt.title('Pages per Languages')

labels = {
    'en': 'English',
    'ja': 'Japanese',
    'de': 'German',
    'na': 'Media',
    'fr': 'French',
    'zh': 'Chinese',
    'ru': 'Russian',
    'es': 'Spanish'
}

for key in sums:
    plt.plot(days, sums[key], label=labels[key])

plt.legend()
plt.show()
```

## 随机统计一下词条的热度

```python

def plot_entry(key, idx):
    fig = plt.figure(figsize=(13,10)) 
    data = lang_sets[key].iloc[idx,1:]
    plt.plot(days, data)
    plt.xlabel('Days')
    plt.ylabel('Views')
    plt.title(str(train.iloc[lang_sets[key].index[idx],0]).decode('utf-8'))
  
idx = [1,20,50,90,100,5000]
for i in idx:
    plot_entry('en', i)
```

## 统计每个国家搜索的热度

```python
for key in lang_sets:
    print('\n\n*******Country: %s*******\n'% labels[key])
    sum_set = pd.DataFrame(lang_sets[key].Page)
    sum_set['total'] = lang_sets[key].sum(axis=1)
    sum_set = sum_set.sort_values(by='total', ascending=False)
    print sum_set.head(10)
```
